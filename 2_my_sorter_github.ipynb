# Comapre Function

```
#@title Convert to spike_interface rec { form-width: "300px" }
def get_spikeinterface_rec_sort(rec_spikeforest, sort_spikeforest):
  time_series, geom, samp_frequency = get_recording_properties(rec_spikeforest)
  spike_times, labels = get_sorting_properties(sort_spikeforest)

  si_rec = se.NumpyRecordingExtractor(timeseries=time_series, geom=geom, sampling_frequency=samp_frequency)
  si_sort = se.NumpySortingExtractor()
  si_sort.set_times_labels(times=spike_times, labels=labels)
  si_sort.set_sampling_frequency(sampling_frequency=samp_frequency)

  return si_rec, si_sort

def get_recording_properties(rec):
    time_series = rec.get_traces()
    geom = rec.get_channel_locations()
    samp_frequency = rec.get_sampling_frequency()

    return time_series, geom, samp_frequency

def get_sorting_properties(sorting):
  spike_train = sorting.get_units_spike_train()

  times_labels = [(t, v + 1) for v in range(len(spike_train)) for t in spike_train[v]]

  times_sorted = sorted(times_labels)

  times = np.array([t for t, label in times_sorted])
  labels = np.array([label for t, label in times_sorted])

  return times, labels
```

```
#@title get_partial_train(rec, sort, sel_size) { form-width: "300px" }
def get_partial_train(rec, sort, sel_size):

  if sel_size == 1:
    return get_spikeinterface_rec_sort(rec, sort)

  else:
    time_series, geom, samp_frequency = get_recording_properties(rec)
    spike_times, labels = get_sorting_properties(sort)

    sel_times_series = []

    sel_spike_times = []
    sel_labels = []

    sel_index = int(len(spike_times) * sel_size)
    sel_spike_times = spike_times[:sel_index]
    sel_labels = labels[:sel_index]

    last_frame = spike_times[sel_index]

    for channel in time_series:
      sel_times_series.append(channel[: last_frame])

    sel_times_series = np.array(sel_times_series)

    si_rec = se.NumpyRecordingExtractor(timeseries=sel_times_series, geom=geom, sampling_frequency=samp_frequency)
    si_sort = se.NumpySortingExtractor()
    si_sort.set_times_labels(times=sel_spike_times, labels=sel_labels)
    si_sort.set_sampling_frequency(sampling_frequency=samp_frequency)

    return si_rec, si_sort
```

```
#@title compare_results { form-width: "300px" }

def compare_results(recurding, sorting_turo, sorters, well_score = 0.7, rep = 20, select_size = [1]):

  '''
  :param here:
    sorters: a list of sorters name; can be ['hdsort', 'herdingspikes', 'ironclust', 'kilosort', 'kilosort2', 'klusta', 
    'mountainsort4', 'spykingcircus', 'tridesclous', 'waveclus']
    well_score: the threshold for a unit to be considered as well-detected
    rep: the number of required repeats
    is_toy: is the given reocrding a toy

  :returns:
    A dataframe with the name of the sorters at rows, and accuracy, precision, recall and #well_detected at columns
  '''

  df = pd.DataFrame(columns=['Mean Acc', 'Std Acc', 'Mean Recall', 'Std Recall',
                             'Mean Precision', 'Std Precision', '# well detected', 'Std # well detected'])
  for cur in sorters:
    for sel_size in select_size:

      rec, sorting_turu = get_partial_train(recurding, sorting_turo, sel_size)
      results_cur = np.empty(shape=(rep, 4))

      for i in range(rep):
        sorting_cur = ss.run_sorter(sorter_name_or_class=cur, recording=rec)
        comp_gt = sc.compare_sorter_to_ground_truth(gt_sorting=sorting_turu, tested_sorting=sorting_cur)
        df_result = comp_gt.get_performance()

        # Accuracy
        results_cur[i, 0] = np.mean(df_result.iloc[:, 0])

        # recall
        results_cur[i, 1] = np.mean(df_result.iloc[:, 1])

        # Precision
        results_cur[i, 2] = np.mean(df_result.iloc[:, 2])

        # n_well detected
        results_cur[i, 3] = comp_gt.count_well_detected_units(well_score)
      
      df.loc[cur + f', train_size: {sel_size}', 'Mean Acc'] = np.mean(results_cur[:, 0])
      df.loc[cur + f', train_size: {sel_size}', 'Std Acc'] = np.std(results_cur[:, 0])

      df.loc[cur + f', train_size: {sel_size}', 'Mean Recall'] = np.mean(results_cur[:, 1])
      df.loc[cur + f', train_size: {sel_size}', 'Std Recall'] = np.std(results_cur[:, 1])

      df.loc[cur + f', train_size: {sel_size}', 'Mean Precision'] = np.mean(results_cur[:, 2])
      df.loc[cur + f', train_size: {sel_size}', 'Std Precision'] = np.std(results_cur[:, 2])

      df.loc[cur + f', train_size: {sel_size}', '# well detected'] = np.mean(results_cur[:, 3])
      df.loc[cur + f', train_size: {sel_size}', 'Std # well detected'] = np.std(results_cur[:, 3])


  return df
```

```
recording, sorting_true = se.example_datasets.toy_example(num_channels=1, duration=4000, seed=0)
df_test = compare_results(recording, sorting_true, ['waveclus'], well_score = 0.75, rep = 20, select_size = [1])
df_test.to_csv('/content/drive/My Drive/Spike Sorting/results/toy_waveclust.csv')
```
